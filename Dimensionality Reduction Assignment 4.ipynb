
# PCA and Clustering Analysis Report

## Introduction

This report presents the results of our analysis, which involved two main components: Principal Component Analysis (PCA) for dimensionality reduction and K-means clustering for grouping data based on similar characteristics. The goal was to gain valuable insights and enhance data understanding.

## PCA (Principal Component Analysis)

### Dimensionality Reduction

We successfully applied PCA to reduce the dimensionality of the dataset. Initially, we started with 13 features, and through PCA, we reduced them to 4 principal components. The choice of the number of principal components was determined through careful analysis.

### Selection of `n_components`

To determine the optimal number of principal components (`n_components`), we utilized a cumulative explained variance plot. This plot visualizes the cumulative contribution of each principal component to the total variance in the data. The plot helped us identify an inflection point where the cumulative explained variance leveled off. Accordingly, we selected 4 as the value for `n_components`, as this point marked the stage at which additional components contributed relatively little to the overall variance.

## Clustering Analysis

### K-means Clustering

For clustering analysis, we utilized the K-means algorithm, a widely-used unsupervised learning technique. We configured the algorithm with the `init` parameter set to 'k-means++' to ensure that the initial cluster centers were distributed effectively, helping to avoid local optima.

### Determining the Number of Clusters (k)

Selecting the appropriate number of clusters, denoted as 'k', is a critical decision in K-means clustering. To make an informed choice, we followed a two-step process:

1. Elbow Method: We visually inspected the within-cluster sum of squares (WCSS) for different values of 'k' (ranging from 1 to 10). The "elbow" point in the plot indicated a diminishing return of WCSS reduction after a certain number of clusters. This method suggested that 3 clusters might be an optimal choice.

2. Kneed Library: We applied the Kneed library to determine the optimal 'k' value programmatically. The Kneed library confirmed that the most suitable number of clusters was indeed 3.

### Resulting Clusters

Based on our analysis, we formed 3 clusters, which align with the inherent structure of the dataset. This outcome corresponds to the fact that the dataset originally includes data on 3 distinct types of wine.

## Conclusion

Our PCA analysis, guided by the cumulative explained variance plot, effectively reduced dimensionality while retaining relevant information. The K-means clustering analysis, with the number of clusters determined through a combination of the Elbow Method and the Kneed Library, produced 3 distinct clusters, aligning with the dataset's inherent structure. This analysis enhances our understanding of the data, enabling more precise data-driven decisions.
